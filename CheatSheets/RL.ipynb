# Reinforcement Learning (RL) — Summary (Jupyter Markdown)

## 1) What is RL?

Reinforcement Learning (RL) trains an **agent** to make **sequential decisions** by interacting with an **environment** and learning from **rewards**.

At time $t$:
- Observe state $S_t$
- Choose action $A_t$
- Receive reward $R_{t+1}$ and next state $S_{t+1}$

Goal: maximize long-run (possibly delayed) reward.

---

## 2) Markov Decision Process (MDP)

Most RL problems are modeled as an MDP:

$$
(\mathcal{S}, \mathcal{A}, P, R, \gamma)
$$

- $\mathcal{S}$: set of states  
- $\mathcal{A}$: set of actions  
- $P(s' \mid s,a)$: transition dynamics  
- $R$ (or $R_{t+1}$): reward function  
- $\gamma \in [0,1)$: discount factor  

**Markov property**:

$$
P(S_{t+1}\mid S_t,A_t,S_{t-1},A_{t-1},\dots)=P(S_{t+1}\mid S_t,A_t)
$$

---

## 3) Return (discounted cumulative reward)

The **return** from time $t$ is:

$$
G_t = \sum_{k=0}^{\infty}\gamma^k R_{t+k+1}
$$

Objective: find a **policy** $\pi(a\mid s)$ that maximizes expected return:

$$
\pi^* = \arg\max_{\pi}\; \mathbb{E}_{\pi}[G_0]
$$

---

## 4) Value Functions

### State-value function
Expected return from state $s$ under policy $\pi$:

$$
V^\pi(s) = \mathbb{E}_\pi[G_t \mid S_t=s]
$$

### Action-value function (Q-function)
Expected return from state $s$, taking action $a$, then following $\pi$:

$$
Q^\pi(s,a) = \mathbb{E}_\pi[G_t \mid S_t=s, A_t=a]
$$

---

## 5) Bellman Equations (policy evaluation)

### Bellman expectation equation for $V^\pi$
$$
V^\pi(s) = \mathbb{E}_\pi\left[ R_{t+1} + \gamma V^\pi(S_{t+1}) \mid S_t=s \right]
$$

### Bellman expectation equation for $Q^\pi$
$$
Q^\pi(s,a) = \mathbb{E}\left[ R_{t+1} + \gamma \mathbb{E}_{a'\sim \pi(\cdot\mid S_{t+1})}\left[Q^\pi(S_{t+1},a')\right] \mid S_t=s, A_t=a \right]
$$

---

## 6) Optimality

Define optimal value functions:

$$
V^*(s)=\max_{\pi}V^\pi(s), \qquad Q^*(s,a)=\max_{\pi}Q^\pi(s,a)
$$

### Bellman optimality equation for $V^*$
$$
V^*(s)=\max_{a}\; \mathbb{E}\left[ R_{t+1}+\gamma V^*(S_{t+1})\mid S_t=s,A_t=a \right]
$$

### Bellman optimality equation for $Q^*$
$$
Q^*(s,a)=\mathbb{E}\left[ R_{t+1}+\gamma \max_{a'}Q^*(S_{t+1},a') \mid S_t=s,A_t=a \right]
$$

Extract an optimal policy:

$$
\pi^*(s)\in \arg\max_{a} Q^*(s,a)
$$

---

## 7) Three Major RL Approach Families

### (A) Dynamic Programming (model known)
Assumes $P$ and $R$ are known; compute $V^\pi$ / $Q^\pi$ and improve policy (e.g., policy iteration, value iteration).

---

### (B) Value-Based RL (model-free)
Learn $Q$ from experience via **Temporal Difference (TD)** learning.

**Q-learning update (off-policy)**:

$$
Q(S_t,A_t)\leftarrow Q(S_t,A_t) + \alpha\Big(R_{t+1}+\gamma\max_{a'}Q(S_{t+1},a')-Q(S_t,A_t)\Big)
$$

The TD error is:

$$
\delta_t = R_{t+1}+\gamma\max_{a'}Q(S_{t+1},a')-Q(S_t,A_t)
$$

Deep version: approximate $Q(s,a)$ with a neural network (e.g., DQN).

---

### (C) Policy-Based RL
Directly optimize a parameterized policy $\pi_\theta(a\mid s)$:

$$
J(\theta)=\mathbb{E}_{\pi_\theta}[G_0]
$$

Policy gradient form:

$$
\nabla_\theta J(\theta)=\mathbb{E}\left[\nabla_\theta \log \pi_\theta(A_t\mid S_t)\; \hat{A}_t \right]
$$

where $\hat{A}_t$ is an advantage estimate.

---

## 8) Actor–Critic (hybrid)
Combine:
- **Actor**: policy $\pi_\theta$
- **Critic**: value function $V_\phi$ or $Q_\phi$

Critic helps estimate $\hat{A}_t$; actor updates policy using it.

---

## 9) Exploration vs Exploitation

A common exploration strategy for value-based methods is $\epsilon$-greedy:

$$
A_t=
\begin{cases}
\text{random action} & \text{with prob } \epsilon \\
\arg\max_a Q(S_t,a) & \text{with prob } 1-\epsilon
\end{cases}
$$

---

## 10) On-policy vs Off-policy

- **On-policy**: learn about the policy used to act (e.g., SARSA, PPO).
- **Off-policy**: learn about a target policy while acting differently (e.g., Q-learning).

---

## 11) Discount Factor $\gamma$

- Smaller $\gamma$: prioritize near-term rewards  
- Larger $\gamma$: prioritize long-term rewards  

---

## 12) When RL works well vs poorly

**Works well for:**
- sequential decisions, delayed outcomes, planning under uncertainty

**Can struggle with:**
- sparse/poorly designed rewards
- non-stationary environments
- low sample-efficiency settings (real-world data is expensive)
- strict safety constraints
