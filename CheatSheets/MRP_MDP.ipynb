# MRP and MDP Cheat Sheet

## 1) Big picture: modeling sequential decision problems

We model a system that evolves over time:

- State at time $t$: $S_t$
- (Maybe) you choose an action $A_t$
- You receive a reward $R_{t+1}$
- The system transitions to a new state $S_{t+1}$

### Markov property
The future depends only on the present (state, and action if applicable):

- **No actions (Markov chain / MRP):**  
  $$\Pr(S_{t+1}=s' \mid S_t=s, S_{t-1},\dots)=\Pr(S_{t+1}=s' \mid S_t=s).$$

- **With actions (MDP):**  
  $$\Pr(S_{t+1}=s' \mid S_t=s, A_t=a, \text{history})=\Pr(S_{t+1}=s' \mid S_t=s, A_t=a).$$

---

## 2) MRP: Markov Reward Process (no decisions)

An **MRP** is a Markov chain plus rewards.

An MRP is specified by:
1. States $\mathcal{S}$
2. Transition probabilities $P(s,s')=\Pr(S_{t+1}=s' \mid S_t=s)$
3. Reward model (often expected reward):
   $$r(s)=\mathbb{E}[R_{t+1}\mid S_t=s]$$
   (or $r(s,s')$ if reward depends on the transition)
4. Discount factor $\gamma \in [0,1)$ (or episodic with termination)

---

## 3) Returns (what value is measuring)

The **return** from time $t$ is the discounted sum of future rewards:

$$G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots.$$

This is **all future steps**, not just one step (unless the process terminates).

---

## 4) Value function in an MRP

The **state-value function** is:

$$v(s) = \mathbb{E}[G_t \mid S_t=s].$$

Interpretation: If I start in state $s$, what total discounted reward should I expect over the future?

---

## 5) Bellman equation for an MRP

Use the identity:

$$G_t = R_{t+1} + \gamma G_{t+1}.$$

Then:

$$v(s)=\mathbb{E}[G_t\mid S_t=s]
=\mathbb{E}[R_{t+1}+\gamma G_{t+1}\mid S_t=s].$$

So:

$$v(s)=\mathbb{E}[R_{t+1}\mid S_t=s]+\gamma\,\mathbb{E}[G_{t+1}\mid S_t=s].$$

Conditioning on next state $S_{t+1}=s'$ gives the **Bellman expectation equation**:

$$v(s)= r(s) + \gamma \sum_{s'} P(s,s')\,v(s').$$

### Matrix form
Let $v$ be the vector of values, $r$ the reward vector, $P$ the transition matrix:

$$v = r + \gamma P v.$$

Rearrange:

$$(I-\gamma P)v = r,$$

so (when the inverse exists):

$$v = (I-\gamma P)^{-1}r.$$

---

## 6) MDP: Markov Decision Process (decisions/actions)

An **MDP** adds control (actions).

An MDP is specified by:
1. States $\mathcal{S}$
2. Actions $\mathcal{A}$
3. Transition dynamics:
   $$P(s'\mid s,a)=\Pr(S_{t+1}=s' \mid S_t=s, A_t=a)$$
4. Reward model:
   $$r(s,a)=\mathbb{E}[R_{t+1}\mid S_t=s, A_t=a]$$
   (or $r(s,a,s')$)
5. Discount factor $\gamma$

**MDP = MRP + actions (control).**

---

## 7) Policies

A **policy** $\pi$ tells you how to choose actions.

- Stochastic policy:
  $$\pi(a\mid s)=\Pr(A_t=a\mid S_t=s).$$

- Deterministic policy:
  $$a=\pi(s).$$

If you **fix a policy** $\pi$, then the MDP becomes an MRP (because actions are chosen according to $\pi$).

---

## 8) Value functions in an MDP (under a policy)

### State-value under $\pi$
$$v_\pi(s)=\mathbb{E}_\pi[G_t\mid S_t=s].$$

### Action-value under $\pi$ (Q-function)
$$q_\pi(s,a)=\mathbb{E}_\pi[G_t\mid S_t=s, A_t=a].$$

Relationship:
$$v_\pi(s)=\sum_a \pi(a\mid s)\,q_\pi(s,a).$$

---

## 9) Bellman expectation equations in an MDP

### For $v_\pi$
$$v_\pi(s)=\sum_a \pi(a\mid s)\Big(r(s,a)+\gamma\sum_{s'}P(s'\mid s,a)\,v_\pi(s')\Big).$$

### For $q_\pi$
$$q_\pi(s,a)=r(s,a)+\gamma\sum_{s'}P(s'\mid s,a)\sum_{a'}\pi(a'\mid s')\,q_\pi(s',a').$$

These are "expectation" equations because they average over:
- action randomness from $\pi$
- transition randomness from $P$

---

## 10) Optimal values and optimal policy

Define the optimal value functions:

$$v_*(s)=\max_\pi v_\pi(s), \quad q_*(s,a)=\max_\pi q_\pi(s,a).$$

An optimal policy $\pi_*$ achieves these maxima.

---

## 11) Bellman optimality equations

### Optimal state-value
$$v_*(s)=\max_a\Big(r(s,a)+\gamma\sum_{s'}P(s'\mid s,a)\,v_*(s')\Big).$$

### Optimal action-value
$$q_*(s,a)=r(s,a)+\gamma\sum_{s'}P(s'\mid s,a)\max_{a'}q_*(s',a').$$

Greedy optimal policy from $q_*$:
$$\pi_*(s)\in\arg\max_a q_*(s,a).$$

---

## 12) Planning algorithms (when you know the MDP)

- **Policy evaluation:** solve for $v_\pi$
- **Policy improvement:** make $\pi$ greedy w.r.t. $v_\pi$
- **Policy iteration:** repeat evaluation + improvement
- **Value iteration:** repeatedly apply Bellman optimality backup:
  $$v_{k+1}(s)\leftarrow \max_a\Big(r(s,a)+\gamma\sum_{s'}P(s'\mid s,a)\,v_k(s')\Big).$$

---

## 13) Direct clarification

### Is the value function one step or all future steps?
It is **all future steps** (the expected discounted return):

$$v(s)=\mathbb{E}\Big[\sum_{k=0}^\infty \gamma^k R_{t+k+1}\,\Big|\,S_t=s\Big].$$

### Is it the same for the Bellman equation?
Yes. The Bellman equation is a recursive rewrite of the same all-future return using:

$$G_t = R_{t+1} + \gamma G_{t+1}.$$

---

# Tiny Examples

## Part A: Tiny MRP example (no decisions)

### Setup
States: $A$, $B$, Terminal $T$ (episode ends at $T$)  
Discount: $\gamma=0.9$

Transitions + rewards:
- From $A$:
  - with prob $0.5$: go to $A$, reward $0$
  - with prob $0.5$: go to $B$, reward $0$
- From $B$:
  - with prob $1$: go to $T$, reward $1$
- Terminal: $v(T)=0$

---

### Step 1: Write Bellman equations

Terminal:
$$v(T)=0.$$

State $B$:
$$v(B)=1+\gamma v(T)=1+0.9\cdot 0=1.$$

State $A$:
$$v(A)=0+\gamma\left(0.5\,v(A)+0.5\,v(B)\right).$$

Plug in $\gamma=0.9$ and $v(B)=1$:
$$v(A)=0.9\left(0.5\,v(A)+0.5\cdot 1\right)=0.45\,v(A)+0.45.$$

Solve:
$$v(A)-0.45v(A)=0.45,$$
$$0.55v(A)=0.45,$$
$$v(A)=\frac{0.45}{0.55}=\frac{45}{55}=\frac{9}{11}\approx 0.81818.$$

**Final MRP values:**
$$v(T)=0,\quad v(B)=1,\quad v(A)\approx 0.81818.$$

---

### (Optional) Value iteration view
Initialize $v_0(A)=0$, $v_0(B)=0$.

Backup rules:
- $$v_{k+1}(B)=1$$
- $$v_{k+1}(A)=0.9\left(0.5v_k(A)+0.5v_k(B)\right)$$

A few iterations:
- $k=0$: $v_0(A)=0,\ v_0(B)=0$
- $k=1$: $v_1(B)=1,\ v_1(A)=0$
- $k=2$: $v_2(A)=0.9(0.5\cdot 0+0.5\cdot 1)=0.45$
- $k=3$: $v_3(A)=0.9(0.5\cdot 0.45+0.5\cdot 1)=0.6525$
- $k=4$: $v_4(A)=0.9(0.5\cdot 0.6525+0.5\cdot 1)=0.743625$

This converges toward $0.81818$.

---

## Part B: Tiny MDP example (with decisions)

### Setup
States: $A$, $B$, Terminal $T$  
Discount: $\gamma=0.9$

Actions (only at $A$):
- **Left:** go to $B$ with reward $0$
- **Right:** go to $T$ with reward $0.4$

At $B$ (no choice):
- go to $T$ with reward $1$

Terminal: $v(T)=0$

---

### Step 1: Solve values

Terminal:
$$v(T)=0.$$

State $B$:
$$v(B)=1+\gamma v(T)=1+0.9\cdot 0=1.$$

At $A$, use **Bellman optimality** (max over actions).

Compute action values:

If Left:
$$Q(A,\text{Left})=0+\gamma v(B)=0.9\cdot 1=0.9.$$

If Right:
$$Q(A,\text{Right})=0.4+\gamma v(T)=0.4+0=0.4.$$

So:
$$v_*(A)=\max(0.9,0.4)=0.9.$$

**Optimal policy:**
- At $A$: choose **Left**
- At $B$: forced transition to $T$

Interpretation: the immediate $0.4$ is worse than taking Left to reach reward $1$ next step, discounted to $0.9$.
